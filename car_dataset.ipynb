```python
# pip package ucimlrepo required
from ucimlrepo import fetch_ucirepo
car_evaluation = fetch_ucirepo(id=19)
```

```python
X = car_evaluation.data.features
X.head()
```

```python
y = car_evaluation.data.targets
y.head()
```

```python
print(car_evaluation.variables)
```

```python
import pandas as pd
# convert the columns to categorical datatype
df_cat = X.astype("category")
y = y.astype("category")
```

```python
# convert categories to their codes
for col in df_cat.columns:
    df_cat[col] = df_cat[col].cat.codes

y["class"] = y["class"].cat.codes

# combine coded X and Y
new_df = pd.concat([df_cat, y], axis=1)
new_df
```

```python
# write to csv
new_df.to_csv("data/car.csv", index=False)
```

```python
max(new_df.buying)
```

```python
domain = {}
for col in new_df.columns:
    domain[col] = max(new_df[col])+1
```

```python
import json
with open("data/car-domain.json", "w") as domain_json:
    json.dump(domain,domain_json)
```
